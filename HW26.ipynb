```python
csv_df = pd.read_csv(r"hw_work.csv")
print("head",csv_df.head())
csv_df.drop(columns=['Area code'], axis=1, inplace=True)
print()
print("row 3332", csv_df.iloc[3332])
csv_df.reset_index(drop=True, inplace=True)
csv_df.drop(index=[3332], axis=0, inplace=True) 
```

```python
csv_df['Total min'] = csv_df['Total night minutes'] + csv_df['Total intl minutes']
print(csv_df['Total min'])
print()
csv_df['Total mean min'] = csv_df[['Total night minutes', 'Total intl minutes']].mean(axis=1)
print(csv_df['Total mean min'])
```

```python
pd_head = csv_df.head(1)
pd_tail = csv_df.tail(1)
new_pd = pd.concat([pd_head, pd_tail], ignore_index=True)
print("new_pd",new_pd)
row = {'Account  length': 23.0, 'International plan': 'Yes', 'Voice mail plan': 'No', 'Number vmail messages': 52, 'Total day minutes': 3030, 'Total day calls': 666,\
        'Total day charge': 3, 'Total eve minutes': 4, 'Total eve calls': 666, 'Total eve charge': 1212, 'Total night minutes':356032, 'Total night calls': 5555,\
          'Total night charge': 1111, 'Total intl minutes': 3333, 'Total intl calls': 66666, 'Total intl charge': 88888, 'Customer service calls': 2222,\
           'Churn': False, 'Total min': 555, 'Total mean min': 222}
my_df_new = pd.DataFrame([row])
print()
print("my df new", my_df_new)
pd.concat([csv_df, pd.DataFrame([row])], ignore_index=True).tail()
pd.concat([csv_df, my_df_new], ignore_index=True).tail()
```
